```python
# ex_15
def group_files(input_files, output_file):
    try:
        with open(output_file, 'w') as ofile:
            for file in input_files:
                with open(file, 'r') as ifile:
                    content = ifile.read()
                    ofile.write(content)
                    ofile.write("\n")
    except Exception as e:
        print("Error occurred:", e)

    
```

```python
input_files = ["input_ex_15_1.txt", "input_ex_15_2.txt", "input_ex_15_3.txt"]  
output_file = "input_ex_15.txt"  
group_files(input_files, output_file)
```
